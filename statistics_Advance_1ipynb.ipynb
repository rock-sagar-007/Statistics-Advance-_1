# Statistics Advance 1


1. What is a random variable in probability theory?
A random variable is a numerical outcome of a random process or experiment. It assigns values to each possible outcome in a sample space.


2. What are the types of random variables?
The two main types are:

Discrete random variables: Take countable values (e.g., number of heads in coin tosses).

Continuous random variables: Take any value within a given range (e.g., height, weight).



3. What is the difference between discrete and continuous distributions?

Discrete distributions deal with countable outcomes (e.g., binomial, Poisson).

Continuous distributions involve uncountably infinite outcomes over an interval (e.g., normal, uniform).



4. What are probability distribution functions (PDF)?

For discrete variables, PDF gives the probability of each outcome (also called PMF — Probability Mass Function).

For continuous variables, the PDF represents the density of the probability; the area under the curve over an interval gives the probability.



5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
The CDF gives the probability that a random variable is less than or equal to a certain value, while the PDF gives the probability of exact values (discrete) or densities (continuous).


6. What is a discrete uniform distribution?
It’s a distribution where all outcomes are equally likely within a finite set (e.g., rolling a fair die).


7. What are the key properties of a Bernoulli distribution?

Only two possible outcomes: success (1) and failure (0).

One trial only.

Probability of success is denoted by p, and failure is 1 - p.



8. What is the binomial distribution, and how is it used in probability?
It describes the number of successes in n independent Bernoulli trials with the same probability p. It’s used in scenarios like flipping coins or testing products.


9. What is the Poisson distribution and where is it applied?
It models the number of events occurring in a fixed interval of time or space, given a known average rate. It's used in modeling rare events like system failures or incoming calls.


10. What is a continuous uniform distribution?
A distribution where all intervals of the same length in the range are equally probable (e.g., picking a random number between 0 and 1).


11. What are the characteristics of a normal distribution?



Bell-shaped and symmetric.

Mean = Median = Mode.

Defined by mean (μ) and standard deviation (σ).


12. What is the standard normal distribution, and why is it important?
A special case of the normal distribution with mean = 0 and standard deviation = 1. It allows easy calculation of probabilities using Z-scores.


13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population’s distribution. It’s crucial for making inferences.


14. How does the Central Limit Theorem relate to the normal distribution?
It explains why normal distribution appears so frequently — it arises naturally from averages of samples, even if the underlying data isn't normally distributed.


15. What is the application of Z statistics in hypothesis testing?
Z statistics are used to determine how far a sample mean is from the population mean in terms of standard deviations. It helps test hypotheses under the normal distribution.


16. How do you calculate a Z-score, and what does it represent?



Z = \frac{X - \mu}{\sigma}

17. What are point estimates and interval estimates in statistics?



Point estimate: A single value estimate of a population parameter (e.g., sample mean).

Interval estimate: A range of values within which the parameter lies, with a certain confidence level.


18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals provide a range that likely contains the population parameter, giving more information than a single point estimate.


19. What is the relationship between a Z-score and a confidence interval?
Z-scores are used to calculate confidence intervals in normal distributions. For example, a 95% confidence interval uses Z = ±1.96.


20. How are Z-scores used to compare different distributions?
Z-scores standardize different data sets, allowing comparisons across distributions with different means and variances.


21. What are the assumptions for applying the Central Limit Theorem?



Random and independent sampling.

A sufficiently large sample size (usually n ≥ 30).

The population should have a finite mean and variance.


22. What is the concept of expected value in a probability distribution?
It is the long-run average value of a random variable, weighted by probabilities:



E[X] = \sum (x_i \cdot P(x_i))

23. How does a probability distribution relate to the expected outcome of a random variable?
The probability distribution assigns likelihoods to outcomes, and the expected value summarizes the average outcome over many trials.
